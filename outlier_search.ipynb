```python
from pybaseball import batting_stats_range
import pandas as pd
```

```python
stats = batting_stats_range("2023-03-30", "2023-05-30")
```

```python
season_stats = batting_stats_range("2023-03-30", "2023-10-01")
```

```python
filtered_stats = stats[stats['PA'] > 100]
filtered_season_stats = season_stats[season_stats['PA'] > 100]
filtered_stats.rename(columns={'OPS':'Early OPS'}, inplace=True)
```

```python
filtered_stats.sort_values(by='Early OPS', ascending=False)
```

```python
stat_df = filtered_season_stats.merge(filtered_stats[['Name', 'Early OPS']], on='Name', how='left')
```

```python
stat_df = stat_df.dropna(subset=['Early OPS'])
```

```python
stat_df['OPS Diff'] = abs(stat_df['OPS'] - stat_df['Early OPS'])
```

```python
stat_df.sort_values(by='OPS Diff', ascending=False).to_csv('sorted_stats.csv')
```
